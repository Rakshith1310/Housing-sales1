```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Canada.csv has year and income from the 1970 till 2016
df=pd.read_csv("E:\KARTHIK\Canada.csv")
print(df.head(3))
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("Per capita income in dollars")
plt.scatter(df.year,df.per_capita_income,color="red",marker="+")
plt.plot(df.year,df.per_capita_income,color="blue")
```

```python
year=df.drop("per_capita_income", axis="columns")
per_capita_income=df.drop("year",axis="columns")
#print(year)
#print(per_capita_income)
```

```python
#fitting the linear regression model
reg=linear_model.LinearRegression()
reg.fit(year,per_capita_income) 
```

```python
#predicting the per_capita_income on the year 2020
reg.predict([[2020]])
```

```python
#"Book1.csv" only has the coloumn "year" from 2017 to 2035
df1=pd.read_csv("E:\KARTHIK\Book1.csv")
print(df1.head(3))
```

```python
prediction=reg.predict(df1)
print(len(prediction))
```

```python
# added another coloumn "prediction" to the dataframe and predicting per_capita_income
df1["prediction"]=prediction
print(df1.head(4))
```

```python
#here "Final_prediction.csv" has the original dataset as well as predicted dataset from 1970-2035
df2=pd.read_csv("E:\KARTHIK\Final_prediction.csv")
print(len(df2))
```
